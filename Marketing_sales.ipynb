```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r"C:\Users\User\Downloads\archive\marketing_sales_dataset.csv")
data =pd.DataFrame(data)
```

```python
data.dropna(inplace=True)
```

```python
data.duplicated()
data.drop_duplicates()
```

```python
data['date']=pd.to_datetime(data['date'])
data.info()
```

```python
data['year']= data.date.dt.year
revenue_by_year = data.groupby(['year','season'])['sales_revenue_usd'].sum().reset_index()
print(revenue_by_year)
sns.lineplot(x = revenue_by_year.season,y=revenue_by_year.sales_revenue_usd,hue=revenue_by_year.year)
plt.title("seasonality")
plt.xlabel("Revenue")
```

```python
sales_of_prouduct_cat =data.groupby(['product_category','season'])['sales_revenue_usd'].sum().reset_index()
print(sales_of_prouduct_cat)
plt.bar(x=sales_of_prouduct_cat.product_category,height=sales_of_prouduct_cat.sales_revenue_usd)
plt.title("Product Category")
plt.ylabel("Rvenue")
plt.xticks(rotation=45)
plt.show()
sns.lineplot(x=sales_of_prouduct_cat.season,y=sales_of_prouduct_cat.sales_revenue_usd,hue=sales_of_prouduct_cat.product_category)
plt.title("seasonality of product categories")
plt.ylable("Revenue")
```

```python
def segment_age(x):
    if x <= 12:
        return '5-12'          
    elif x <= 19:
        return '13-19'       
    elif x <= 35:
        return '20-35'    
    elif x <= 55:
        return '36-55'          
    elif x <= 75:
        return '56-75'    
    else:
        return 'senior'         

data['age_segment'] = data['customer_age'].apply(segment_age)
age_segment_group= data['age_segment'].value_counts()
print(age_segment_group)
plt.pie(age_segment_group,labels=age_segment_group.index,autopct ='%1.1f%%')
```

```python
sns.barplot(x=data.age_segment,y=data.sales_revenue_usd,hue=data.customer_segment)
```

```python
sns.barplot(x = data.customer_segment, y=data.sales_revenue_usd)
#sns.barplot(x = data.customer_segment, y=data.num_previous_purchases)
plt.show()
```

```python
def segment_satisfaction(x):
    if x <= 2:
        return 'dissatisfied'    
    elif x == 3:
        return 'neutral'          
    elif x == 4:
        return 'satisfied'        
    else:
        return 'very satisfied'
customer_satisfiction_segment = data.customer_satisfaction_score.apply(segment_satisfaction)
customer_satisfiction_segment_group = customer_satisfiction_segment.value_counts()
plt.pie(customer_satisfiction_segment_group,labels= customer_satisfiction_segment_group.index,autopct='%1.1f%%')
```
